## STEP -01   IMPORT REQUIRED PYTHON PACKAGES.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

## STEP -02   LOAD THE DATASETS

```python
data=pd.read_csv('crime.csv')
data
```

## STEP -03   DATA ANALYSIS USING PANDAS

```python
data.info()
```

```python
data.columns
```

```python
# Checking for missing values
print(data.isnull().sum())

# Summary statistics
print(data.describe())

# Viewing the first few rows of the DataFrame
print(data.head())
```

## STEP -04   SPLIT DATA SET INTO INDEPENDENT/DEPENDENT VARAIBLE

```python
# Step 04: Split Dataset into Independent/Dependent Variables
# We will predict the total recorded crime for Pakistan using other columns.

# Independent variables
X = data[['Year', 'Punjab', 'Sindh', 'KP', 'Balochistan', 'Islamabad', 'Railways', 'G.B', 'AJK']]

# Dependent variable
y = data['Pakistan']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Step 05: Train Regression Model

```python

# Creating the regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## STEP -06   PREDICT THE RESULT 

```python
### Step 06: Predict the Result
# Predicting the results for the test set
y_pred_test = regressor.predict(X_test)
y_pred_train = regressor.predict(X_train)
```

## Step 07: Plot the Training and Test Results

```python
# Step 07: Plot the Training and Test Results
# Plot actual vs predicted values for the test set
plt.scatter(X_test['Year'], y_test, color='lightcoral', label='Actual Test Values')
plt.scatter(X_test['Year'], y_pred_test, color='blue', marker='x', label='Predicted Test Values')
plt.title('Total Recorded Crime vs Year (Test Set)')
plt.xlabel('Year')
plt.ylabel('Total Recorded Crime in Pakistan')
plt.legend()
plt.box(False)
plt.show()

# Plot the regression line (Train set)
plt.scatter(X_train['Year'], y_train, color='lightcoral', label='Actual Train Values')
plt.plot(X_train['Year'], y_pred_train, color='firebrick', linewidth=2, label='Regression Line')
plt.title('Total Recorded Crime vs Year (Train Set)')
plt.xlabel('Year')
plt.ylabel('Total Recorded Crime in Pakistan')
plt.legend()
plt.box(False)
plt.show()

# Printing the regressor coefficients and intercept
print(f'Coefficients: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```
